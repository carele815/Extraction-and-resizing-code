**Extract the images from the different folders situated in the folder "GOOD"** 

```python
  
destination_path = "/content/drive/MyDrive/PlumUnzinveau/Prune/extracted_nveau_good/"
pattern = "/content/drive/MyDrive/PlumUnzinveau/Prune/GOOD/*/*"  
for img in glob.glob(pattern):
    shutil.copy(img, destination_path)
```

**Now let's resize them**

```python

for img in os.listdir('extracted_nveau_good'):
    open_image  = Image.open(destination_path + img)
    new_img = open_image.resize((256,256))
    os.chdir('/content/drive/MyDrive/PlumUnzinveau/Prune/resized_nveau/good_resized_nveau')
    new_img.save(img)
      #print(img)
    os.chdir('/content/drive/MyDrive/PlumUnzinveau/Prune/resized_nveau')
```
